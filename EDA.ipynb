```python
# Machine learning model training and evaluation
```

```python
# Data preprocessing
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
data=['gender', 'race/ethnicity', 'parental level of education', 'lunch',
       'test preparation course']
for i in data:
    dataset[i]=encoder.fit_transform(dataset[i].astype(str))
dataset
```

```python
# Training the model to find the individual scores
```

```python
x=dataset.drop(columns=['avg score','writing score']).values
X=dataset.drop(columns=['avg score','writing score'])
x1=dataset.drop(columns=['avg score','math score']).values
X1=dataset.drop(columns=['avg score','math score'])
x2=dataset.drop(columns=['avg score','reading score']).values
X2=dataset.drop(columns=['avg score','reading score'])
```

```python
y=dataset['writing score'].values
y1=dataset['math score'].values
y2=dataset['reading score'].values
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0) 
x1_train, x1_test, y1_train, y1_test= train_test_split(x1, y1, test_size=0.25, random_state=0)
x2_train, x2_test, y2_train, y2_test= train_test_split(x2, y2, test_size=0.25, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
model1=LinearRegression()
model2=LinearRegression()
model3=LinearRegression()
```

```python
model1.fit(x_train,y_train)
model2.fit(x1_train,y1_train)
model3.fit(x2_train,y2_train)
```

```python
p1=model1.predict(x_test)
p2=model2.predict(x1_test)
p3=model3.predict(x2_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print("Accuracy score of writing score : ",r2_score(p1,y_test))
```

```python
print("Accuracy score of math score : ",r2_score(p2,y1_test))
```

```python
print("Accuracy score of reading score : ",r2_score(p3,y2_test))
```
